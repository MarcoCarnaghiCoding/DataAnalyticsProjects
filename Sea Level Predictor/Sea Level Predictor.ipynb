Using matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
y = df['CSIRO Adjusted Sea Level']
x = df['Year']

fig, ax = plt.subplots()
plt.scatter(x,y)
```

Using the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 

Then, plotting the line of best fit over the top of the scatter plot. 

Also, Making the line go through the year 2050 to predict the sea level rise in 2050.

```python
from scipy.stats import linregress
best_fit_line = linregress(x,y)
x_pred = pd.Series([i for i in range(1880,2050)])
y_pred = best_fit_line.slope * (x_pred) + best_fit_line.intercept

plt.plot(x_pred, y_pred, 'r--')

ax.set_title('Rise in Sea Level')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
```

```python
df = pd.read_csv(filepath)

y = df['CSIRO Adjusted Sea Level']
x = df['Year']


# Create scatter plot

fig, ax = plt.subplots()
plt.scatter(x,y)

# Create first line of best fit
best_fit_line = linregress(x,y)
x_pred = pd.Series([i for i in range(1880,2050)])
y_pred = best_fit_line.slope * (x_pred) + best_fit_line.intercept

plt.plot(x_pred, y_pred, 'r--')

# Create second line of best fit
df2 = df.loc[df['Year']>=2000]
x2 = df2['Year']
y2 = df2['CSIRO Adjusted Sea Level']

res2 = linregress(x2,y2)
x_pred2 = pd.Series([i for i in range(2000,2050)])
y_pred2 = res2.slope * (x_pred2) + res2.intercept

plt.plot(x_pred2, y_pred2, 'g--')
# Add labels and title
ax.set_title('Rise in Sea Level')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
```
